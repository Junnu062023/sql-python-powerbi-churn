```python
import pandas as pd
import matplotlib.pyplot as plt

# Load CSV again
df = pd.read_csv("C:/Users/manda/Documents/Churn_Modelling.csv")
df.columns = df.columns.str.strip()

# Group by Tenure and calculate churn rate
tenure_churn = df.groupby('Tenure')['Exited'].mean().reset_index()
tenure_churn['Exited'] = tenure_churn['Exited'] * 100

# Plot the line chart
fig, ax = plt.subplots(figsize=(10, 6))
ax.plot(tenure_churn['Tenure'], tenure_churn['Exited'], marker='o', color='darkgreen')
ax.set_title("Churn Rate by Tenure")
ax.set_xlabel("Tenure (Years)")
ax.set_ylabel("Churn Rate (%)")
ax.grid(True)

# FORCE DISPLAY
plt.show(block=True)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 5))
sns.countplot(x='NumOfProducts', hue='Exited', data=df, palette='viridis')
plt.title("Churn Count by Number of Products")
plt.xlabel("Number of Products")
plt.ylabel("Customer Count")
plt.legend(title='Churned (Exited)', labels=['No', 'Yes'])
plt.show()
```

```python
plt.figure(figsize=(6, 4))
sns.countplot(x='IsActiveMember', hue='Exited', data=df, palette='coolwarm')
plt.title("Churn Count by Active Membership")
plt.xlabel("Active Member (1 = Yes, 0 = No)")
plt.ylabel("Customer Count")
plt.legend(title='Churned', labels=['No', 'Yes'])
plt.show()
```

```python
plt.figure(figsize=(6, 4))
sns.countplot(x='HasCrCard', hue='Exited', data=df, palette='Set2')
plt.title("Churn Count by Credit Card Ownership")
plt.xlabel("Has Credit Card (1 = Yes, 0 = No)")
plt.ylabel("Customer Count")
plt.legend(title='Churned', labels=['No', 'Yes'])
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(
    x='Age',
    y='Balance',
    hue='Exited',
    data=df,
    palette='Set1',
    alpha=0.6
)
plt.title("Customer Balance vs Age (Colored by Churn)")
plt.xlabel("Age")
plt.ylabel("Balance")
plt.legend(title="Churned", labels=["No", "Yes"])
plt.grid(True)
plt.show()
```
